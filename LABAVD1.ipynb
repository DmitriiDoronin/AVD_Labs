```python
#Introduction data in the matrix
X_matrix = [[1, -1, 2],
            [2, 0, 0],
            [0, 1, -1]]
X_matrix
```

```python
#mean computing
average = np.average(X_matrix)
average
```

```python
#variance computing
variance = np.var(X_matrix)
variance
```

```python
#Import and normalization of data with MinMax scaler
from sklearn.preprocessing import MinMaxScaler
#minmax scaler
#xi = (x-min(x))/(max(x)-min(x)) 
#--->[0,1]
MinMax = MinMaxScaler()
X_matrix1 = MinMax.fit_transform(X_matrix)
X_matrix1
```

```python
#Import and normalization of data with Standart scaler
from sklearn.preprocessing import StandardScaler
#Standartization
#z = (x-mean(x)/sigma)
#z - score mean = 0, standart deviation = 1
StScaler = StandardScaler()
X_matrix2 = StScaler.fit_transform(X_matrix)
X_matrix2
```

```python
#Computing mean and averagr afte MinMaxscaler
average1 = np.average(X_matrix1)
variance1 = np.var(X_matrix1)
print('Mean after MinMaxScaler = ' ,round(average1, 2))
print('Variance after MinMaxScaler = ', round(variance1, 2))
```

```python
#Computing mean and averagr afte StandartScaler
average2 = np.average(X_matrix2)
variance2 = np.var(X_matrix2)
print('Mean after MinMaxScaler = ' ,round(average2, 2))
print('Variance after MinMaxScaler = ', round(variance2, 2))
```

```python
#Training data
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
X = iris.loc[:, features].values
```

```python
#Importing PCA module from sklearn and training on Iris data with 3 components
from sklearn.decomposition import PCA

X = StandardScaler().fit_transform(X)
pca = PCA(n_components=3)
Y = iris.loc[:,['Species']].values
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2','principal component 3'])
finalDf = pd.concat([principalDf, iris[['Species']]], axis = 1)
finalDf.head(7)
```

```python
pca.explained_variance_ratio_.round(2)
```

```python
fig = plt.figure()
fig.set_figheight(19)
fig.set_figwidth(23)
fig.subplots_adjust(hspace=0.5, wspace=0.5)

axe = fig.add_subplot(3, 3, 1)
xaxe = list(range(-75,75,1))
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
for i in range(len(finalDf['principal component 1'])):
    plt.scatter(xaxe[i],finalDf['principal component 1'][i], c=colors[finalDf['Species'][i]])
plt.xlabel("Index")
plt.ylabel("principal component 1")

axe = fig.add_subplot(3, 3, 2)
for i in range(len(finalDf['principal component 1'])):
    plt.scatter(xaxe[i],finalDf['principal component 2'][i], c=colors[finalDf['Species'][i]])
plt.xlabel("Index")
plt.ylabel("principal component 2")

axe = fig.add_subplot(3, 3, 3)
for i in range(len(finalDf['principal component 1'])):
    plt.scatter(xaxe[i],finalDf['principal component 3'][i], c=colors[finalDf['Species'][i]])
plt.xlabel("Index")
plt.ylabel("principal component 3")

axe = fig.add_subplot(3, 3, 5)
for i in range(len(finalDf['principal component 1'])):
    plt.scatter(finalDf['principal component 1'][i], finalDf['principal component 2'][i],c=colors[finalDf['Species'][i]])
plt.xlabel("principal component 1")
plt.ylabel("principal component 2")
```

```python
#Importing LDA module from sklearn and training on Iris data with 2 components
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
X= iris.loc[:, features].values
X = StandardScaler().fit_transform(X)
Y = iris.loc[:,['Species']].values
Y  = np.ravel(Y,order='A')
lda = LDA(n_components=2)
X_train = lda.fit_transform(X, Y)
LDA_Df = pd.DataFrame(data = X_train
             , columns = ['component 1', 'component 2'])
final_LDA_Df = pd.concat([LDA_Df, iris[['Species']]], axis = 1)
```

```python
# Importing and visualising MNIST dataset
from matplotlib import pyplot
from scipy.io import loadmat

mnist = loadmat("mnist-original.mat")
mnist_data = mnist["data"].T
mnist_label = mnist["label"][0]

def load_mnist_dataset():

    mnist = loadmat("mnist-original.mat")
    mnist_data = mnist["data"].T
    mnist_data = mnist_data.reshape(len(mnist_data), 28, 28, 1)
    mnist_label = mnist["label"][0]
    count = len(set(mnist_label))
    return mnist_data, mnist_label, count, (28, 28, 1)
# Download data from MNIST-original dataset
data, labels, classes_count, data_shape = load_mnist_dataset()
print("data shape: " + str(data.shape))
print("labels shape: " + str(labels.shape))
print("classes count: " + str(classes_count))
def plot_images_sample(X, Y):
    # Draw plot for images sample
    
    plt.figure(figsize=(10,10))
    rand_indicies = np.random.randint(len(X), size=25)
    for i in range(25):
        plt.subplot(5,5,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        index = rand_indicies[i]
        plt.imshow(np.squeeze(X[index]), cmap=plt.cm.binary)
        plt.xlabel(Y[index])
    plt.show()
# Draw plot for images sample
plot_images_sample(data, labels)
```
